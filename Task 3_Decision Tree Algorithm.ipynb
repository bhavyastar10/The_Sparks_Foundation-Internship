## 1. Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn import metrics
import sklearn.datasets as datasets
```

## 2. Importing Dataset

```python
df= pd.read_csv("Iris.csv")
df.head(10)
```

```python
df.drop(['Id'], axis=1, inplace=True) #removing column "id" from dataframe
df.info()
```

```python
df.shape  #to know about number of rows and columns
```

```python
df.describe()
```

```python
df.corr()  #relation between each columns
```

## 3. Pairplot for visualization of dataset

```python
sns.pairplot(df, hue="Species")  #visual co-relation between columnns using inbuilt function of seaborn
```

## 4. Preparing dataset

```python
x = df.drop("Species", axis=1).to_numpy()  #converting dataframe to numpy array
x
```

```python
y_text= df['Species'].to_numpy()
y = LabelEncoder().fit_transform(y_text)
y
```

## 5. Spliting dataset for Training and Testing dataset

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.3, random_state=0)
```

## 6. Applying Decision Tree classifier 

```python
dtree= DecisionTreeClassifier()  #initialising desicion tree classifier
dtree.fit(x_train, y_train)
```

## 7. Finding the accuracy 

```python
yhat_test= dtree.predict(x_test)   #predicting y values using decision tree classifier
yhat_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, yhat_test)  #finding accuracy of prediction
```

## 8. Visualizing the Decision Tree 

```python
plt.figure(figsize=(20,10))
# forming decision tree from out 
tree_img=plot_tree(dtree,feature_names=df.columns,class_names=df['Species'].unique().tolist(),precision=4,label="all",filled=True)
plt.show()
```
