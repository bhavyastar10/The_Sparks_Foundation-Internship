## 1. Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

## 2. Importing data 

```python
iris= datasets.load_iris()
iris_df= pd.DataFrame(iris.data, columns= iris.feature_names)
iris_df.head(10)
```

```python
# to get information of dataframe
iris_df.info()
```

```python
iris_df.describe()
```

```python
# get to know about relation between each columns
iris_df.corr()
```

## 3. Finding optimum number of clusters

```python
# converting dataframe to numpy array
x = iris_df.to_numpy()
x
```

```python
wcss= []
# making visualize graph for different values of K
for i in range (1,21):
    model=KMeans(i)
    model.fit(x)
    wcss.append(model.inertia_)
plt.plot(range(1,21), wcss)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

## 4. Visualizing the clusters

```python
model=KMeans(3)
# because from above graph , we can say that K=3 is break point
y_kmeans=model.fit_predict(x)
fig=plt.figure(figsize=(10,5))
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'blue', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'green', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'red', label = 'Iris-virginica')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:,1],s = 100, c = 'black', label = 'Centroids')
plt.legend()
```

### <li>Conclusion: </li>
Although as k increases , accuracy also increases. but we take minimum value of K , where graph is changing abruptly it's direction. 
We can see intersection between blue and red clusters , but when we compare it with K=2, it is far better than that result.
